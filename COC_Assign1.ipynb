```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.linear_model import LinearRegression
from scipy import stats

df = pd.read_csv("/Student_Marks.csv", encoding='cp1252')
print(df)
```

```python
plt.scatter(df['number_courses'],df['Marks'])
plt.xlabel('Number Of Courses')
plt.ylabel('Marks')
```

```python
plt.scatter(df['time_study'],df['Marks'])
plt.xlabel('Time of study')
plt.ylabel('Marks')
```

```python
slope, intercept, rValue, pv, se = stats.linregress(df['time_study'],df['Marks'])
sn.regplot(x='time_study',y='Marks',ci = None, data= df,line_kws = {"color":"green"}, label="y={0:.1f}x+{1:.1f}".format(slope,intercept)).legend(loc="best")
```

```python
Average_Marks = df['Marks'].mean()
print(Average_Marks)

df['Performance']= df['Marks']-Average_Marks
print(df)
```
